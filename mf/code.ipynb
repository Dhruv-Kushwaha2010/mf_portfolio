```python
from docx import Document
import pandas as pd

# Load .docx file
doc = Document("AMFI_Reports.docx")

# Select first table (change index for different tables)
table = doc.tables[0]

# Extract data
data = []
for row in table.rows:
    data.append([cell.text.strip() for cell in row.cells])

# Convert to DataFrame
df = pd.DataFrame(data)


# rename columns and drop first 5 rows
df.columns = df.iloc[4].values
df = df[5:]
df.index = df['NAV date']
df.index.name = 'date'
df.index = pd.to_datetime(df.index)
df = df.iloc[:, :1]
display(df)
```

```python
px.line(df, x=df.index, y='Net Asset Value', title='Net Asset Value')
```
